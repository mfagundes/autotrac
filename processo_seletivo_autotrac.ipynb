## 21.Mostre um código em Pandas que tem como objetivo a criação de uma série a partir de um dicionário

```python
d = {"A": 100, "B": 200, "C":300}
s1 = pd.Series(d, name="Nome da série")
```

```python
s1
```

```python
# É possível alterar a ordem dos índices e, ainda, incluir um índice inexistente no dicionário,
# que terá como valor atribuído np.nan

s2 = pd.Series(d, name="Série com NaN", index=['C', 'A', 'B', 'D'])
```

```python
s2
```

Obs: como a inserção do índice D, o valor np.nan inserido nele fez com que o Pandas alterasse o dtype para float64. Observe que a anterior, como continha apenas inteiros, foi gerada com dtype int64

### c. Como seria o código para ordenar de maneira ascendente o DataSet pelas colunas ‘Estado’ e ‘Recuperados’?


#### Ordenação por "Estado" e "Recuperado"
Obs: Como a questão pede a ordenação pelas colunas "Estado" e "Recuperado", entendo que a odenação seria por ambas, nessa ordem. Porém há apenas uma linha para cada estado, assim o DataFrame ordenado por ambas as colunas é idêntico ao que seria ordenado apenas por Estado


```python
df_c1 = df.sort_values(['Estado', 'Recuperado'], ascending=True)
```

```python
df_c1
```

#### Ordenação por "Recuperado" e "Estado"
Ordenando primeiro por "Recuperado", agrupamos as linhas com o mesmo número de recuperados e, assim, estas ficam em ordem ascendente de "Estado"

```python
df_c2 = df.sort_values(['Recuperado', 'Estado'], ascending=True)
```

```python
df_c2
```
